## Introducción al Plotting con Matplotlib y Seaborn

una gráfica bien diseñada puede revelar patrones, tendencias y valores atípicos que serían imposibles de detectar directamente en una tabla de datos rugosos. 

En este notebook aprenderemos a construir visualizaciones interactivas y estáticas desde una hoja en blanco. Utilizaremos el ecosistema estándar de Python:
* **Matplotlib:** Nuestra base y lienzo, que nos da el control absoluto de cada elemento.
* **Seaborn:** Una capa de alto nivel que simplifica la creación de gráficos estadísticos hermosos con pocas líneas de código.

A lo largo de esta guía práctica, exploraremos cómo estructurar un gráfico correctamente y resolveremos un caso de estudio real con el dataset de los **Taxis de Nueva York**.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings               # <--- 1. Importas la librería de alertas

# 2. Le dices a Python que ignore los "FutureWarning"
warnings.filterwarnings("ignore", category=FutureWarning)

# Configuración de estilo
sns.set_style("whitegrid")
taxis_df = sns.load_dataset("taxis")
```

# ¿Cómo se distribuyen las distancias de los viajes?
Histograma (histplot)

```python
fig = plt.figure(figsize=(10, 6))
sns.histplot(data=taxis_df, x="distance", kde=True, bins=30)
plt.title("Distribution of Trip Distances")
plt.xlabel("Distance (miles)")
plt.ylabel("Frequency")
plt.show()
```

# Cuál es la relación entre la distancia del viaje y la tarifa? 
Gráfico de dispersión (scatterplot).

```python
fig = plt.figure(figsize=(10, 6))
sns.scatterplot(data=taxis_df, x="distance", y="fare", alpha=0.6)
plt.title("Relationship between Distance and Fare")
plt.xlabel("Distance (miles)")
plt.ylabel("Fare (USD)")
plt.show()
```

# ¿Cuál es el promedio de las propinas por cada zona de recogida?

```python
# Filtramos las top 10 zonas con nombres de variables
top_pickup_zones = taxis_df.groupby("pickup_zone")["tip"].mean().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=top_pickup_zones, x="tip", y="pickup_zone", palette="Blues_r")
plt.title("Top 10 Pickup Zones by Average Tip")
plt.xlabel("Average Tip (USD)")
plt.ylabel("Pickup Zone")
plt.show()
```

# ¿Cuál es la cantidad de pasajeros por barrio (borough) de recogida?

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=taxis_df, x="pickup_borough", y="passengers", estimator=sum, errorbar=None, palette="viridis")
plt.title("Total Passengers by Pickup Borough")
plt.xlabel("Pickup Borough")
plt.ylabel("Total Passengers")
plt.show()
```

# ¿Hay alguna relación entre el número de pasajeros y la tarifa total?
Gráfico de caja (boxplot) o de líneas (lineplot)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=taxis_df, x="passengers", y="total")
plt.title("Total Fare Distribution by Number of Passengers")
plt.xlabel("Number of Passengers")
plt.ylabel("Total Fare (USD)")
plt.show()
```

# ¿Qué métodos de pago (payment) son los más comunes?
Gráfico de conteo (countplot).

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=taxis_df, x="payment", palette="Set2")
plt.title("Most Common Payment Methods")
plt.xlabel("Payment Method")
plt.ylabel("Trip Count")
plt.show()
```

# Cómo varían las tarifas según el barrio de recogida?
Diagrama de caja (boxplot) para ver la dispersión y los outliers.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=taxis_df, x="pickup_borough", y="fare")
plt.title("Fare Variation by Pickup Borough")
plt.xlabel("Pickup Borough")
plt.ylabel("Fare (USD)")
plt.show()
```

# Qué método de pago recibe las mayores propinas?
Diagrama de violín (violinplot) o de barras (barplot) para comparar los promedios.

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=taxis_df, x="payment", y="tip", palette="pastel", errorbar=None)
plt.title("Average Tip by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Average Tip (USD)")
plt.show()
```

# Hay diferencias en las tarifas totales dependiendo del color del taxi?
Diagrama de caja (boxplot) o histogramas superpuestos.

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=taxis_df, x="color", y="total", palette=["#F7D117", "#4CBB17"])
plt.title("Total Fare by Taxi Color")
plt.xlabel("Taxi Color")
plt.ylabel("Total Fare (USD)")
plt.show()
```

# ¿Cuánto se paga en peajes según el borough de destino?
Gráfico de barras (barplot) mostrando el promedio o el total acumulado de peajes (tolls).

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=taxis_df, x="dropoff_borough", y="tolls", estimator=sum, errorbar=None, palette="magma")
plt.title("Total Tolls Paid by Dropoff Borough")
plt.xlabel("Dropoff Borough")
plt.ylabel("Total Tolls (USD)")
plt.show()
```

# ¿Cuál es la distancia media de los viajes según el borough de destino?
Gráfico de barras (barplot) calculando la media de forma automática.

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=taxis_df, x="dropoff_borough", y="distance", palette="coolwarm", errorbar=None)
plt.title("Average Trip Distance by Dropoff Borough")
plt.xlabel("Dropoff Borough")
plt.ylabel("Average Distance (miles)")
plt.show()
```
